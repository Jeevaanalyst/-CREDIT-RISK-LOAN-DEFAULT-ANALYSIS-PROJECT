```python
# =============================================================================
# CREDIT RISK & LOAN DEFAULT ANALYSIS
# DATA ANALYST PROJECT
# =============================================================================

import pandas as pd
import numpy as np

# =============================================================================
# LOAD DATA
# =============================================================================

df = pd.read_csv("loan_default_dataset_1lakh.csv")


print("CREDIT RISK & LOAN DEFAULT ANALYSIS")


print("\nDataset Shape:", df.shape)


```

```python
# DATA CLEANING
# =============================================================================

print("Dataset Shape:", df.shape)

# Remove duplicates
df = df.drop_duplicates()

# Handle missing values
df.fillna({
    "annual_income": df["annual_income"].median(),
    "credit_score": df["credit_score"].median(),
    "employment_type": "Unknown"
}, inplace=True)
```

```python
# =============================================================================
# DATA QUALITY CHECKS
# =============================================================================
print("\nMISSING VALUES")
print(df.isnull().sum())


print("\nDUPLICATE RECORDS")
print(df.duplicated().sum())
```

```python
# =============================================================================
# PORTFOLIO KPI SUMMARY
# =============================================================================

portfolio_kpis = pd.DataFrame({

    "Metric":[
        "Total Loans",
        "Portfolio Value",
        "Average Loan Amount",
        "Average Credit Score",
        "Default Rate (%)",
        "Average Interest Rate (%)"
    ],

    "Value":[
        len(df),
        round(df["loan_amount"].sum(),0),
        round(df["loan_amount"].mean(),0),
        round(df["credit_score"].mean(),0),
        round(df["default"].mean()*100,2),
        round(df["interest_rate"].mean(),2)
    ]
})

print("\nPORTFOLIO KPI SUMMARY")

display(portfolio_kpis)
```

```python
# =============================================================================
# EXPLORATORY DATA ANALYSIS 

#LOAN GRADE ANALYSIS
# =============================================================================

loan_grade_analysis = (

    df.groupby("loan_grade")

    .agg(
        Applications=("loan_id","count"),
        Default_Rate=("default","mean"),
        Avg_Loan_Amount=("loan_amount","mean"),
        Avg_Credit_Score=("credit_score","mean")
    )

    .reset_index()
)

loan_grade_analysis["Default_Rate"] = (
    loan_grade_analysis["Default_Rate"]*100
).round(2)

print("\nLOAN GRADE ANALYSIS")
display(loan_grade_analysis)
```

```python
# =============================================================================
# EMPLOYMENT TYPE ANALYSIS
# =============================================================================

employment_analysis = (

    df.groupby("employment_type")

    .agg(
        Applications=("loan_id","count"),
        Default_Rate=("default","mean"),
        Avg_Income=("annual_income","mean")
    )

    .reset_index()
)

employment_analysis["Default_Rate"] = (
    employment_analysis["Default_Rate"]*100
).round(2)

print("\nEMPLOYMENT TYPE ANALYSIS")
display(employment_analysis)
```

```python
# LOAN PURPOSE ANALYSIS
# =============================================================================

purpose_analysis = (

    df.groupby("loan_purpose")

    .agg(
        Applications=("loan_id","count"),
        Default_Rate=("default","mean"),
        Avg_Loan=("loan_amount","mean")
    )

    .reset_index()
)

purpose_analysis["Default_Rate"] = (
    purpose_analysis["Default_Rate"]*100
).round(2)

print("\nLOAN PURPOSE ANALYSIS")
display(purpose_analysis)
```

```python
# CREDIT SCORE BAND ANALYSIS
# =============================================================================

df["credit_band"] = pd.cut(

    df["credit_score"],

    bins=[300,550,650,750,850],

    labels=[
        "Poor",
        "Fair",
        "Good",
        "Excellent"
    ]
)

credit_band_analysis = (

    df.groupby("credit_band")

    .agg(
        Applications=("loan_id","count"),
        Default_Rate=("default","mean"),
        Avg_Loan=("loan_amount","mean")
    )

    .reset_index()
)
credit_band_analysis["Default_Rate"] = (
    credit_band_analysis["Default_Rate"]*100
).round(2)

print("\nCREDIT BAND ANALYSIS")
display(credit_band_analysis)
```

```python
# =============================================================================
# STATE LEVEL ANALYSIS
# =============================================================================

state_analysis = (

    df.groupby("state")

    .agg(
        Applications=("loan_id","count"),
        Default_Rate=("default","mean"),
        Avg_Loan=("loan_amount","mean"),
        Avg_Credit_Score=("credit_score","mean")
    )

    .reset_index()
)

state_analysis["Default_Rate"] = (
    state_analysis["Default_Rate"]*100
).round(2)

state_analysis = state_analysis.sort_values(
    by="Default_Rate",
    ascending=False
)

top_states = state_analysis.head(10)

print("\nTOP 10 STATES BY DEFAULT RATE")
display(top_states)
```

```python
# DEBT TO INCOME ANALYSIS
# =============================================================================

df["DTI_Band"] = pd.qcut(
    df["debt_to_income"],
    5
)

dti_analysis = (

    df.groupby("DTI_Band")

    .agg(
        Applications=("loan_id","count"),
        Default_Rate=("default","mean")
    )

    .reset_index()
)

dti_analysis["Default_Rate"] = (
    dti_analysis["Default_Rate"]*100
).round(2)

print("\nDEBT TO INCOME ANALYSIS")
display(dti_analysis)
```

```python
# =============================================================================
# RISK TIER SEGMENTATION
# =============================================================================

df["Risk_Tier"] = pd.cut(

    df["credit_score"],

    bins=[300,550,650,750,850],

    labels=[
        "High Risk",
        "Medium Risk",
        "Low Risk",
        "Very Low Risk"
    ]
)

risk_tier_analysis = (

    df.groupby("Risk_Tier")

    .agg(
        Applications=("loan_id","count"),
        Default_Rate=("default","mean"),
        Avg_Loan=("loan_amount","mean"),
        Avg_Income=("annual_income","mean")
    )

    .reset_index()
)
risk_tier_analysis["Default_Rate"] = (
    risk_tier_analysis["Default_Rate"]*100
).round(2)

print("\nRISK TIER ANALYSIS")
display(risk_tier_analysis)
```

```python
high_risk_customers = (

    df[
        (df["credit_score"] < 600) &
        (df["debt_to_income"] > 0.75)
    ]

    .sort_values(
        by="loan_amount",
        ascending=False
    )

    .head(100)
)

print("\nHIGH RISK CUSTOMER ")
display(high_risk_customers.head())
```

```python
# EXPORT REPORT TO EXCEL
# =============================================================================

output_file = "Credit_Risk_Analysis_Report.xlsx"

with pd.ExcelWriter(output_file) as writer:

    portfolio_kpis.to_excel(
        writer,
        sheet_name="Portfolio KPI",
        index=False
    )

    loan_grade_analysis.to_excel(
        writer,
        sheet_name="Loan Grade",
        index=False
    )

    employment_analysis.to_excel(
        writer,
        sheet_name="Employment Type",
        index=False
    )

    purpose_analysis.to_excel(
        writer,
        sheet_name="Loan Purpose",
        index=False
    )

    credit_band_analysis.to_excel(
        writer,
        sheet_name="Credit Band",
        index=False
    )

    state_analysis.to_excel(
        writer,
        sheet_name="State Analysis",
        index=False
    )

    dti_analysis.to_excel(
        writer,
        sheet_name="DTI Analysis",
        index=False
    )

    risk_tier_analysis.to_excel(
        writer,
        sheet_name="Risk Tier",
        index=False
    )

    high_risk_customers.to_excel(
        writer,
        sheet_name="High Risk Customers",
        index=False
    )

print("\n" + "="*70)
print("REPORT GENERATED SUCCESSFULLY")
display("Output File : Credit_Risk_Analysis_Report.xlsx")
print("="*70)
```
